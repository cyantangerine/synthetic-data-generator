```python
from mycode.multi_ctgan import MultiTableCTGAN
from mycode import dataset
from mycode.testcode import Xargs
db_path = dataset.DatabaseTest.path_1k
db_level = db_path[db_path.rfind("/")+1:db_path.rfind(".db")]
table_level = '14'
x_args = Xargs.XArg(**Xargs.XArgs.tables_14._asdict())
from mycode.dataset import XMetaBuilder
# for i in range(10,21,5):
if True:
    i = 0 # 10, 15, 20
    print(i)
    mmm = MultiTableCTGAN(
        temp_name=f'{db_level}{table_level}t{i}a',
        db_path=db_path,
        x_table=x_args.x_table,
        x_key=x_args.x_key,
        x_how=x_args.x_how
    ).init(add_cols_num=i)
    mmm.save_add_data(f"./mycode/testcode/{mmm.TABLE_TEMP_NAME}.pkl")
    mmm.build_metadata(builder=XMetaBuilder(x_args))
    %memit mmm.fit(device="cpu", batch_size=2000)
    for ss in range(2):
        num = 1000 * (10 ** ss) # 1000, 10000, 100000
        df = mmm.sample(num)
        mmm.evaluate(x_args, df, str(num))
```
